# **Balance**

```python
mean_balance = data['Balance'].mean()
std_balance = data['Balance'].std()
variance_balance = data['Balance'].var()
skewness_balance = data['Balance'].skew()
kurtosis_balance =data['Balance'].kurt()

print(f"Mean: {mean_balance}")
print(f"Standard Deviation: {std_balance}")
print(f"Variance: {variance_balance}")
print(f"Skewness: {skewness_balance}")
print(f"Kurtosis: {kurtosis_balance}")
```

# **credit score**

```python
mean_credit_score = data['CreditScore'].mean()
std_credit_score = data['CreditScore'].std()
variance_credit_score = data['CreditScore'].var()
skewness_credit_score = data['CreditScore'].skew()
kurtosis_credit_score =data['CreditScore'].kurt()

print(f"Mean: {mean_credit_score}")
print(f"Standard Deviation: {std_credit_score}")
print(f"Variance: {variance_credit_score}")
print(f"Skewness: {skewness_credit_score}")
print(f"Kurtosis: {kurtosis_credit_score}")
```

# **transaction count**

```python
mean_transaction_count = data['Balance'].mean()
std_transaction_count = data['Balance'].std()
variance_transaction_count = data['Balance'].var()
skewness_transaction_count = data['Balance'].skew()
kurtosis_transaction_count =data['Balance'].kurt()

print(f"Mean: {mean_transaction_count}")
print(f"Standard Deviation: {std_transaction_count}")
print(f"Variance: {variance_transaction_count}")
print(f"Skewness: {skewness_transaction_count}")
print(f"Kurtosis: {kurtosis_transaction_count}")
```

# **Distribution**

```python
import pandas as pd

# Distribution of customers by gender
gender_distribution = data['Gender'].value_counts(normalize=True) * 100
print("Gender Distribution:\n", gender_distribution)

# Distribution of customers by geography
geography_distribution = data['Geography'].value_counts(normalize=True) * 100
print("\nGeography Distribution:\n", geography_distribution)

# Distribution of customers by age (using bins for better visualization)
age_bins = [0, 20, 30, 40, 50, 60, 70, 80, float('inf')]  # Adjust bins as needed
age_labels = ['0-20', '21-30', '31-40', '41-50', '51-60', '61-70', '71-80', '80+']
data['AgeGroup'] = pd.cut(data['Age'], bins=age_bins, labels=age_labels, right=False)
age_distribution = data['AgeGroup'].value_counts(normalize=True) * 100
print("\nAge Distribution:\n", age_distribution)

# Combined distribution (example: by gender and geography)
combined_distribution = data.groupby(['Gender', 'Geography'])['CustomerId'].count().unstack()
print("\nCombined Gender and Geography Distribution:\n", combined_distribution)
```

```python
transaction_stats = data.groupby('IsActiveMember')['NumOfProducts'].agg(['min', 'max', 'mean'])

transaction_stats
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Correlation Analysis
correlation_matrix = data[['CreditScore', 'Balance', 'NumOfProducts']].corr()
print("\nCorrelation Matrix:\n", correlation_matrix)

# Visualization of relationships
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Credit Score, Balance, and Number of Products')
plt.show()


# Scatter plots to visualize relationships
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
sns.scatterplot(x='CreditScore', y='Balance', data=data)
plt.title('Credit Score vs. Balance')

plt.subplot(1, 2, 2)
sns.scatterplot(x='NumOfProducts', y='Balance', data=data)
plt.title('Number of Products vs. Balance')

plt.tight_layout()
plt.show()

#More detailed analysis with boxplots
plt.figure(figsize=(12, 6))

plt.subplot(1,2,1)
sns.boxplot(x='NumOfProducts', y='Balance', data=data)
plt.title('Balance Distribution by Number of Products')

plt.subplot(1,2,2)
sns.boxplot(x='NumOfProducts', y='CreditScore', data=data)
plt.title('Credit Score Distribution by Number of Products')

plt.tight_layout()
plt.show()
```

```python

churn_stats = data.groupby('Exited')[['Balance', 'NumOfProducts']].agg(['mean', 'median', 'std', 'min', 'max'])

print(churn_stats)

# Visualize the distributions of balance and transaction activity for churned vs. non-churned customers
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.boxplot(x='Exited', y='Balance', data=data)
plt.title('Balance Distribution by Churn Status')

plt.subplot(1, 2, 2)
sns.boxplot(x='Exited', y='NumOfProducts', data=data)
plt.title('Number of Products Distribution by Churn Status')

plt.tight_layout()
plt.show()

#Further analysis with KDE plots
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.kdeplot(data[data['Exited'] == 0]['Balance'], label='Stayed', shade=True)
sns.kdeplot(data[data['Exited'] == 1]['Balance'], label='Churned', shade=True)
plt.title('Balance Distribution (KDE)')
plt.legend()

plt.subplot(1, 2, 2)
sns.kdeplot(data[data['Exited'] == 0]['NumOfProducts'], label='Stayed', shade=True)
sns.kdeplot(data[data['Exited'] == 1]['NumOfProducts'], label='Churned', shade=True)
plt.title('Number of Products Distribution (KDE)')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
low_credit_score_threshold = 500
high_balance_threshold = 10000

low_credit_high_balance_customers = data[
    (data['CreditScore'] < low_credit_score_threshold) & (data['Balance'] > high_balance_threshold)
].shape[0]

total_customers = data.shape[0]

# Calculate the probability
probability = low_credit_high_balance_customers / total_customers

print(f"Probability of low credit score and high balance: {probability}")
```
